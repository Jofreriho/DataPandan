```python
##Calcular autocorrelacion de la serie de tiempo
from statsmodels.graphics.tsaplots import plot_acf

Desde_DfOriginal =0

if (Desde_DfOriginal==0):
##Lo tomaremos desde el Dataframe de Fecha
  df_FechaAgrupada=df_FechaHurtoMotos.groupby(df_FechaHurtoMotos.dt.date).size()
else:
  df_FechaAgrupada=Df_HurtoMotos.groupby(Df_HurtoMotos['fecha_hecho']).size()


#
plt.figure(figsize=(12,6))
plot_acf(df_FechaAgrupada)
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
if (Desde_DfOriginal==0):
  plt.title('Autocorrelation FUNCTION (ACF) PLOT(DESDE CALCULADO FECHAS)')
else:
  plt.title('Autocorrelation FUNCTION (ACF) PLOT (DESDE Original)')

plt.show()

##dfHurtoPersonaDiaSeman= dfHurtoPersona.fecha_hecho.dt.day_of_week
##dfHurtoPersonaDiasemana
```
